
# 各データの特徴量特性把握

### 欠損値を把握

```python
def get_null_stat(source_df):
    '''
    return null count and percentge for each columns of source_df
    '''
    total_count = source_df.shape[0]
    null_count = source_df.isnull().sum().values
    null_ratio = null_count / total_count
    return pd.DataFrame(data={'null_count' : null_count, 'null_ratio' : null_ratio}, index=source_df.columns.values)
```

```python
get_null_stat(train_df).sort_values('null_count', ascending=False).head(70)
```

```python
get_null_stat(test_df).sort_values('null_count', ascending=False).head(70)
```

```python
get_null_stat(prev_application_df).sort_values('null_count', ascending=False).head(20)
```

```python
get_null_stat(bureau_df).sort_values('null_count', ascending=False).head(10)
```

```python
get_null_stat(bureau_balance_df).sort_values('null_count', ascending=False).head(10)
```

```python
get_null_stat(credit_balance_df).sort_values('null_count', ascending=False).head(10)
```

```python
get_null_stat(inst_payments_df).sort_values('null_count', ascending=False).head(5)
```

```python
get_null_stat(pos_cash_df).sort_values('null_count', ascending=False).head(5)
```
